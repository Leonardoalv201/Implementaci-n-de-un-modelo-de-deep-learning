##Importamos las librerias que usaremos

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from keras.models import load_model
import numpy as np
```

Posibles resultados del modelo

```python
ans = {
    0: 'pizza',
    1: 'helado'
}
```

##Carga del modelo

```python
model = load_model('Pizza-Ice_model.h5')
```

##Prediccion del modelo

```python
#Obtencion de la imagen
image = tf.keras.utils.load_img('pizza2.jpeg', target_size=(150, 150))

#Modificaciones previas a la imagen para que tenga el formato necesario
img = np.array(image)
img = img / 255.0
img = img.reshape(1,150,150,3)

#Prediccion del modelo
label = model.predict(img)

#Clasificacion según el resultado
if label[0][0] < 0.5:
  index = 0
else:
  index = 1
```

```python
#Obtencion de la imagen
image2 = tf.keras.utils.load_img('helado1.jpg', target_size=(150, 150))

#Modificaciones previas a la imagen para que tenga el formato necesario
img2 = np.array(image2)
img2 = img2 / 255.0
img2 = img2.reshape(1,150,150,3)

#Prediccion del modelo
label2 = model.predict(img2)

#Clasificacion según el resultado
if label2[0][0] < 0.5:
  index2 = 0
else:
  index2 = 1
```

##Plot visual

Ejemplo 1

```python
import matplotlib.pyplot as plt

plt.imshow(image)
plt.title(ans[index])
```

Ejemplo 2

```python
plt.imshow(image2)
plt.title(ans[index2])
```
